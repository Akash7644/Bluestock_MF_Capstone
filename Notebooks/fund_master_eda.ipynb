# Understanding AMFI Scheme Codes

## Observations

- Each mutual fund scheme is assigned a unique AMFI code.
- Direct and Regular plans of the same fund have different AMFI codes.
  - Example:
    - SBI Bluechip Fund - Regular Plan → 119551
    - SBI Bluechip Fund - Direct Plan → 119552
- AMFI codes are not necessarily sequential across all schemes.
  - Example:
    - HDFC Top 100 Fund - Regular Plan → 100016
    - HDFC Top 100 Fund - Direct Plan → 125497
- The AMFI code serves as a unique identifier for every scheme-plan combination.

## Role in the Project

AMFI codes are used to:

- Link Fund Master and NAV History datasets.
- Validate referential integrity across datasets.
- Fetch scheme data from MFAPI endpoints.
- Support ETL processing, analytics, and dashboard development.

## Conclusion

The AMFI code is the primary key used throughout the Mutual Fund Analytics project. It uniquely identifies each mutual fund scheme and enables reliable integration of data across multiple sources.

# Fund Master Dataset Analysis

## Dataset Overview

- The dataset contains **40 mutual fund schemes** and **15 attributes**.
- Each scheme is uniquely identified using an **AMFI code**.
- The dataset includes important scheme-level information such as fund house, category, sub-category, benchmark, expense ratio, fund manager, and risk category.

## Key Findings

### Fund Houses
- Multiple Asset Management Companies (AMCs) are represented in the dataset.
- Major fund houses include SBI Mutual Fund, HDFC Mutual Fund, ICICI Prudential Mutual Fund, Nippon India Mutual Fund, Axis Mutual Fund, and Kotak Mutual Fund.

### Categories and Sub-Categories
- Schemes are distributed across major categories such as **Equity, Debt, and Hybrid**.
- Sub-categories include investment styles such as **Large Cap, Mid Cap, Small Cap, Flexi Cap, Gilt, Liquid, Aggressive Hybrid**, and others.

### Risk Classification
- The dataset contains schemes across different risk levels:
  - Low
  - Moderate
  - High
  - Very High

### Data Quality Observations
- No missing values were found in the dataset.
- No duplicate records were identified.
- Date fields such as `launch_date` are currently stored as object/string datatype and may require conversion to datetime during ETL processing.

## Conclusion

The Fund Master dataset serves as the primary reference table for the Mutual Fund Analytics project. It provides scheme metadata and AMFI codes that will be used to validate NAV history data and support subsequent ETL, analytics, and dashboard development.
